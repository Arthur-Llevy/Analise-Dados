```python
# uso o "import" para poder trazer as bibliotecas "pandas" e "matplotlib".
# Também uso o "as" para renomeá-las para não precisar escrever o nome completo delas todas as vezes que for usá-las
import pandas as pd
import matplotlib.pyplot as plt

# crio a variável "base_dados" para armazenar os dados da minha planilha em excel.
# uso o método "read_excel" da biblioteca "pandas" para ler meu arquivo e extrair seus dados
base_dados = pd.read_excel('/content/drive/MyDrive/baseEnade17_Sistemas.xlsx')

# crio a variável "quantidade_linhas_plot" para definir quantas linhas de gráficos meu layout terá
quantidade_linhas_plot = 3

# armazeno o quantitativo total de alunos usando "shape[0]", que retorna o número de linhas
quantitativo_alunos_total = base_dados.shape[0]

# armazeno o quantitativo de alunos na modalidade EaD (quando a coluna 'MODALIDADE' for igual a 0)
# para isso, uso o método loc, nele passo a condição que eu quero, nesse caso, quero os registros que tenham o campo modalidade igual a zero
# que representa os alunos da modalidade ead
quantitativo_alunos_ead = base_dados.loc[base_dados['MODALIDADE'] == 0].shape[0]

# armazeno o quantitativo de alunos na modalidade presencial (quando a coluna 'MODALIDADE' for igual a 1)
# para isso, uso o método loc, nele passo a condição que eu quero, nesse caso, quero os registros que tenham o campo modalidade igual a 1
# que representa os alunos da modalidade presencial
quantitativo_alunos_presencial = base_dados.loc[base_dados['MODALIDADE'] == 1].shape[0]

# defino os valores do eixo Y e as categorias do eixo X para o primeiro gráfico
y_grafico_1 = [quantitativo_alunos_ead, quantitativo_alunos_presencial]
x_grafico_1 = ["Ead", "Presencial"]

# crio a variável "tabela_frequencia_1" para poder realizar a contagem dos registros da coluna "MODALIDADE", para isso, uso o "value_counts "
# uso o "reset_index" para poder redefinir o índice da minha tabela, para começar do 0
tabela_frequencia_1 = base_dados["MODALIDADE"].value_counts().reset_index()

# defino o nome das colunas da minha tabela, passando uma lista na propriedade "columns" da minha tabela
tabela_frequencia_1.columns = ['Modalidade', 'Frequência']

# informo que vou exibir a tabela de frequência com o print
# e exibo a tabela de frequência usando o print com a minha variável dentro
print("Tabela de frequência da modalidade de ensino")
print(tabela_frequencia_1)

# defino o tamanho do gráfico utilizando o método "figure" com o parâmetro "figsize"
plt.figure(figsize=(20, 20))

# subdivido o espaço de plotagem, criando um espaço para o primeiro gráfico na posição 1 (linha 1, coluna 1)
plt.subplot(quantidade_linhas_plot, 2, 1)

# crio o gráfico de barras com os dados dos eixos X e Y e defino as cores para cada barra
plt.bar(x_grafico_1, y_grafico_1, color=["blue", "orange"])

# adiciono o rótulo do eixo X
plt.xlabel("Modalidade de ensino")

# adiciono o rótulo do eixo Y
plt.ylabel("Quantitativo de alunos")

# adiciono o título do gráfico
plt.title("Análise por modalidade de ensino")

# adiciono uma grade ao gráfico
plt.grid()

# defino os valores do eixo Y em porcentagem, calculando a proporção em relação ao total de alunos
y_grafico_2 = [(quantitativo_alunos_ead / quantitativo_alunos_total) * 100, (quantitativo_alunos_presencial / quantitativo_alunos_total) * 100]

# defino as categorias do eixo X para o segundo gráfico (representação em porcentagem por modalidade)
x_grafico_2 = ["Ead", "Presencial"]

# subdivido o espaço de plotagem, criando um espaço para o segundo gráfico na posição 2 (linha 1, coluna 2)
plt.subplot(quantidade_linhas_plot, 2, 2)

# crio o gráfico de barras com os dados em porcentagem e defino as cores
plt.bar(x_grafico_2, y_grafico_2, color=["blue", "orange"])

# adiciono o rótulo do eixo X
plt.xlabel("Modalidade de ensino")

# adiciono o rótulo do eixo Y (em porcentagem)
plt.ylabel("Quantitativo de alunos (em porcentagem %)")

# adiciono o título do gráfico
plt.title("Análise por modalidade de ensino (em porcentagem - %) em relação ao total")

# adiciono uma grade ao gráfico
plt.grid()

# armazeno o quantitativo de alunos do sexo masculino na modalidade EaD
quantitativo_alunos_masculino_ead = base_dados.loc[(base_dados['SEXO'] == "M") & (base_dados['MODALIDADE'] == 0)].shape[0]

# armazeno o quantitativo de alunos do sexo feminino na modalidade EaD
quantitativo_alunos_feminino_ead = base_dados.loc[(base_dados['SEXO'] == "F") & (base_dados['MODALIDADE'] == 0)].shape[0]

# defino os valores do eixo Y para o terceiro gráfico (gênero na modalidade EaD)
y_grafico_3 = [quantitativo_alunos_masculino_ead, quantitativo_alunos_feminino_ead]

# defino as categorias do eixo X (gêneros)
x_grafico_3 = ["Masculino", "Feminino"]

# crio a variável "tabela_frequencia_2" para poder realizar a contagem dos registros da coluna "sexo", para isso, uso o "value_counts "
# uso o "reset_index" para poder redefinir o índice da minha tabela, para começar do 0
tabela_frequencia_3 = base_dados.loc[base_dados["MODALIDADE"] == 0, "SEXO"].value_counts().reset_index()

# defino o nome das colunas da minha tabela, passando uma lista na propriedade "columns" da minha tabela
tabela_frequencia_3.columns = ['SEXO', 'Frequência']

# uso esses prints para criar espaçamento entre as tabelas
print("")
print("-----------------------------------------")
print("")

# informo que vou exibir a tabela de frequência com o print
# e exibo a tabela de frequência usando o print com a minha variável dentro
print("Tabela de frequência da coluna sexo na modalidade EaD")
print(tabela_frequencia_3)

# subdivido o espaço de plotagem, criando um espaço para o terceiro gráfico na posição 3 (linha 2, coluna 1)
plt.subplot(quantidade_linhas_plot, 2, 3)

# crio o gráfico de barras para o gênero na modalidade EaD
plt.bar(x_grafico_3, y_grafico_3, color=["coral", "green"])

# adiciono o rótulo do eixo X
plt.xlabel("Gênero")

# adiciono o rótulo do eixo Y
plt.ylabel("Quantitativo")

# adiciono o título do gráfico
plt.title("Divisão de gênero na modalidade ead")

# adiciono uma grade ao gráfico
plt.grid()

# armazeno o quantitativo de alunos do sexo masculino na modalidade presencial
quantitativo_alunos_masculino_presencial = base_dados.loc[(base_dados['SEXO'] == "M") & (base_dados['MODALIDADE'] == 1)].shape[0]

# armazeno o quantitativo de alunos do sexo feminino na modalidade presencial
quantitativo_alunos_feminino_presencial = base_dados.loc[(base_dados['SEXO'] == "F") & (base_dados['MODALIDADE'] == 1)].shape[0]

# defino os valores do eixo Y para o quarto gráfico (gênero na modalidade presencial)
y_grafico_4 = [quantitativo_alunos_masculino_presencial, quantitativo_alunos_feminino_presencial]

# defino as categorias do eixo X (gêneros)
x_grafico_4 = ["Masculino", "Feminino"]

# crio a variável "tabela_frequencia_4" para poder realizar a contagem dos registros da coluna "sexo" onde os alunos são da modalidade presencial,
# para isso, uso o "value_counts "
# uso o "reset_index" para poder redefinir o índice da minha tabela, para começar do 0
tabela_frequencia_4 = base_dados.loc[base_dados["MODALIDADE"] == 1, "SEXO"].value_counts().reset_index()

# defino o nome das colunas da minha tabela, passando uma lista na propriedade "columns" da minha tabela
tabela_frequencia_4.columns = ['SEXO', 'Frequência']

# uso esses prints para criar espaçamento entre as tabelas
print("")
print("-----------------------------------------")
print("")

# informo que vou exibir a tabela de frequência com o print
# e exibo a tabela de frequência usando o print com a minha variável dentro
print("Tabela de frequência da coluna sexo na modalidade presencial")
print(tabela_frequencia_4)

# subdivido o espaço de plotagem, criando um espaço para o quarto gráfico na posição 4 (linha 2, coluna 2)
plt.subplot(quantidade_linhas_plot, 2, 4)

# crio o gráfico de barras para o gênero na modalidade presencial
plt.bar(x_grafico_4, y_grafico_4, color=["coral", "green"])

# adiciono o rótulo do eixo X
plt.xlabel("Gênero")

# adiciono o rótulo do eixo Y
plt.ylabel("Quantitativo")

# adiciono o título do gráfico
plt.title("Divisão de gênero na modalidade presencial")

# adiciono uma grade ao gráfico
plt.grid()

# armazeno o quantitativo de alunos do turno matutino
quantitativo_alunos_turno_matutino = base_dados.loc[base_dados['TURNO'] == 1].shape[0]

# armazeno o quantitativo de alunos do turno vespertino
quantitativo_alunos_turno_vespertino = base_dados.loc[base_dados['TURNO'] == 2].shape[0]

# armazeno o quantitativo de alunos do turno integral
quantitativo_alunos_turno_integral = base_dados.loc[base_dados['TURNO'] == 3].shape[0]

# armazeno o quantitativo de alunos do turno noturno
quantitativo_alunos_turno_noturno = base_dados.loc[base_dados['TURNO'] == 4].shape[0]

# defino os valores do eixo Y para o quinto gráfico (quantitativo por turno)
y_grafico_5 = [quantitativo_alunos_turno_matutino, quantitativo_alunos_turno_vespertino, quantitativo_alunos_turno_integral, quantitativo_alunos_turno_noturno]

# defino as categorias do eixo X (turnos)
x_grafico_5 = ["Matutino", "Vespertino", "Integral", "Noturno"]

# crio a variável "tabela_frequencia_5" para poder realizar a contagem dos registros da coluna "turno"
# para isso, uso o "value_counts "
# uso o "reset_index" para poder redefinir o índice da minha tabela, para começar do 0
tabela_frequencia_5 = base_dados["TURNO"].value_counts().reset_index()

# defino o nome das colunas da minha tabela, passando uma lista na propriedade "columns" da minha tabela
tabela_frequencia_5.columns = ['TURNO', 'Frequência']

# uso esses prints para criar espaçamento entre as tabelas
print("")
print("-----------------------------------------")
print("")

# informo que vou exibir a tabela de frequência com o print
# e exibo a tabela de frequência usando o print com a minha variável dentro
print("Tabela de frequência da coluna turno")
print(tabela_frequencia_5)

# subdivido o espaço de plotagem, criando um espaço para o quinto gráfico na posição 5 (linha 3, coluna 1)
plt.subplot(quantidade_linhas_plot, 2, 5)

# crio o gráfico de barras para o quantitativo por turno
plt.bar(x_grafico_5, y_grafico_5, color=["blue", "green", "firebrick", "darkviolet"])

# adiciono o rótulo do eixo X
plt.xlabel("Turno")

# adiciono o rótulo do eixo Y
plt.ylabel("Quantitativo")

# adiciono o título do gráfico
plt.title("Divisão por turno")

# adiciono uma grade ao gráfico
plt.grid()

# armazeno o quantitativo de alunos da rede pública
quantitativo_alunos_publica = base_dados.loc[base_dados['CategoriaAdm'] == 'Publica'].shape[0]

# armazeno o quantitativo de alunos da rede privada
quantitativo_alunos_privada = base_dados.loc[base_dados['CategoriaAdm'] == 'Privada'].shape[0]

# defino os valores do eixo Y para o sexto gráfico (quantitativo por rede de ensino)
y_grafico_6 = [quantitativo_alunos_publica, quantitativo_alunos_privada]

# defino as categorias do eixo X (redes de ensino)
x_grafico_6 = ["Pública", "Privada"]

# crio a variável "tabela_frequencia_6" para poder realizar a contagem dos registros da coluna "turno"
# para isso, uso o "value_counts "
# uso o "reset_index" para poder redefinir o índice da minha tabela, para começar do 0
tabela_frequencia_5 = base_dados["CategoriaAdm"].value_counts().reset_index()

# defino o nome das colunas da minha tabela, passando uma lista na propriedade "columns" da minha tabela
tabela_frequencia_5.columns = ['CategoriaAdm', 'Frequência']

# uso esses prints para criar espaçamento entre as tabelas
print("")
print("-----------------------------------------")
print("")

# informo que vou exibir a tabela de frequência com o print
# e exibo a tabela de frequência usando o print com a minha variável dentro
print("Tabela de frequência da coluna CategoriaAdm")
print(tabela_frequencia_5)

# subdivido o espaço de plotagem, criando um espaço para o sexto gráfico na posição 6 (linha 3, coluna 2)
plt.subplot(quantidade_linhas_plot, 2, 6)

# crio o gráfico de barras para o quantitativo por rede de ensino
plt.bar(x_grafico_6, y_grafico_6, color=["navy", "gold"])

# adiciono o rótulo do eixo X
plt.xlabel("Rede de ensino")

# adiciono o rótulo do eixo Y
plt.ylabel("Quantitativo")

# adiciono o título do gráfico
plt.title("Divisão por rede de ensino")

# adiciono uma grade ao gráfico
plt.grid()

# ajusto o layout para que os elementos do gráfico fiquem bem organizados
plt.tight_layout()

# exibo todos os gráficos que foram criados
plt.show()

#por fim, uso prints para mostrar os resultados da minha análise. Alguns deles são apenas para dar espaçamento
print("-------------------------------------------------------------------")
print("")
print("Análise dos dados")
print("")
print("MODALIDADE DE ENSINO")
print("Segundo os dados, os alunos que cursam na modalidade presencial representam mais de 80% de todos os alunos")
print("O quantitativo de alunos na modalidade presencial é pouco maior que 5 vezes do que a quantidade dos alunos na modalidade EaD")
print("")

print("-------------------------------------------------------------------")
print("Análise por divisão de gênero")
print("")
print("Na modalidade EaD, o número de estudantes do sexo masculino é pouco mais de oito vezes maior do que o quantitativo de estudantes do sexo feminino")
print("Na modalidade presencial, o quantitativo de estudantes do sexo masculino é pouco mais de 4 vezes maior do que o número de estudantes do sexo feminino")
print("")

print("-------------------------------------------------------------------")
print("Análise por divisão de turno")
print("")
print("O turno noturno é o que mais possui estudantes matriculados, seguido pelo turno matutino, integral e vespertino, respectivamente")
print("Na modalidade presencial, o quantitativo de estudantes do sexo masculino é pouco mais de 4 vezes maior do que o número de estudantes do sexo feminino")
print("")


print("-------------------------------------------------------------------")
print("Análise por divisão de rede de ensino")
print("")
print("A rede pública de ensino tem cerca de 2 vezes mais estudantes matriculados do do que a rede privada de ensino")
print("")
```

## Questão 2 - Escolher duas das variáveis categóricas

```python
# importo as bibliotecas pandas e matplotlib e as renomeio com o "as"

import pandas as pd
import matplotlib.pyplot as plt


# carrego minha base de dados com o método read_excel
base_dados = pd.read_excel('/content/drive/MyDrive/baseEnade17_Sistemas.xlsx')

# crio a tabela de contingência, passando os dados que quero cruzar, nesse caso, as colunas "CategoriaAdm" e  "SEXO"
# passo o atributo margin como True para ele me retornar o somatório
tabela_de_contingencia = pd.crosstab(base_dados["CategoriaAdm"], base_dados["SEXO"], margins=True, margins_name="Somatório")

# uso o plot para por os dados no gráfico. Defino que será um gráfico de barras com o "kind=bar" e mudo as cores das barras
tabela_de_contingencia.plot(kind="bar", color=["green", "firebrick", "black"])

# exibo minha tabela com o print
print(tabela_de_contingencia)

# mudo o rótulo do eixo y e x respectivamente
plt.ylabel("Quantitativo de Estudantes")
plt.xlabel("Rede de ensino")

# altero o título
plt.title("Análise entre as colunas 'sexo' e 'CategoriaAdm'")

# coloco uma grade com o grid
plt.grid()

# e exibo o gráfico com o show
plt.show()


# crio quatro variáveis, para armazenarem o quantitativos de estudantes do sexo masculino e feiminino da rede pública e privada
# para isso, uso o "loc" do pandas, passando duas condições: o sexo - que poder ser masculino ou feminino - e o tipo de rede de ensino
# que é a coluna CategoriaAdm, onde decido se será pública ou privada
# no final, uso o "shape[0]" para pegar todos os registros que atendam a essas condições
estudantes_sexo_masculino_rede_privada = (base_dados.loc[(base_dados["SEXO"] == "M") & (base_dados["CategoriaAdm"] == "Privada")]).shape[0]
estudantes_sexo_masculino_rede_publica = (base_dados.loc[(base_dados["SEXO"] == "M") & (base_dados["CategoriaAdm"] == "Publica")]).shape[0]

estudantes_sexo_feminino_rede_privada = (base_dados.loc[(base_dados["SEXO"] == "F") & (base_dados["CategoriaAdm"] == "Privada")]).shape[0]
estudantes_sexo_feminino_rede_publica = (base_dados.loc[(base_dados["SEXO"] == "F") & (base_dados["CategoriaAdm"] == "Publica")]).shape[0]


# por fim, com os prints, exibo minha análise
# uso o "f" para poder concatenar os valores das variáveis que usei
print("A partir dos dados extraídos da análise, é possível perceber que tanto na rede privada e pública e ensino, estudantes do sexo masculino são a maior parte.")
print("")
print(f"Na rede privada, o quantitativo de estudantes do sexo masculino é {((estudantes_sexo_masculino_rede_privada / estudantes_sexo_feminino_rede_privada)) * 100}% maior se comparado as estudantes do sexo feminino ")
print("")
print(f"Na rede pública, o quantitativo de estudantes do sexo masculino é {round(((estudantes_sexo_masculino_rede_publica / estudantes_sexo_feminino_rede_publica)), 2) * 100}% maior se comparado as estudantes do sexo feminino ")
print("")
print("Concluí-se que, o quantitativo de estudantes do sexo masculino nos dois tipos de rede de ensino (privada e pública) é maior")

```

## Questão 3 - Fazer um estudo descritivo/exploratório


```python
# importo as bibliotecas pandas e matplotlib e as renomeio com o "as"

import pandas as pd
import matplotlib.pyplot as plt


# carrego minha base de dados com o método read_excel
base_dados = pd.read_excel('/content/drive/MyDrive/baseEnade17_Sistemas.xlsx')


# crio quatro varáveis, uma para média, mediana, desvio padrão e coeficiente de variação.
# nas 3 primeiras uso o "loc" pois quero que elas armazenem valores em certas condições. Nesse caso, a modalidade igual a zero
# que representa o ead, após isso, a coluna que quero usar e o que vou fazer com ela

# uso o mean para calcular a média
media_ead_nota_geral = base_dados.loc[base_dados["MODALIDADE"] == 0, "NOTA_GERAL"].mean()

# uso o median para calcular a mediana
mediana_ead_nota_geral = base_dados.loc[base_dados["MODALIDADE"] == 0, "NOTA_GERAL"].median()

# uso o std para calcular o desvio padrão
desvio_padrao_ead_nota_geral = base_dados.loc[base_dados["MODALIDADE"] == 0, "NOTA_GERAL"].std()

# e para calcular o coecificente de variação, pego o desvio padrão, divido pela média e multiplico por 100
coeficiente_variacao_ead = (desvio_padrao_ead_nota_geral / media_ead_nota_geral) * 100


# para as próximas 4 variáveis, fiz a mesma coisa, apenas mudei a condição
media_presencial_nota_geral = base_dados.loc[base_dados["MODALIDADE"] == 1, "NOTA_GERAL"].mean()
mediana_presencial_nota_geral = base_dados.loc[base_dados["MODALIDADE"] == 1, "NOTA_GERAL"].median()
desvio_padrao_presencial_nota_geral = base_dados.loc[base_dados["MODALIDADE"] == 1, "NOTA_GERAL"].std()
coeficiente_variacao_presencial = (desvio_padrao_presencial_nota_geral / media_presencial_nota_geral) * 100


# crio as duas variáveis abaixo para pegar todos os valores do campo nota_geral em que a modalidade seja ead, para a primeira variável
# e presencial para a segunda
notas_ead = base_dados.loc[base_dados["MODALIDADE"] == 0, "NOTA_GERAL"]
notas_presencial = base_dados.loc[base_dados["MODALIDADE"] == 1, "NOTA_GERAL"]


# insiro meus dados no bloxplot. Altero a legenda do eixo x com o parâmetro labels
plt.boxplot([notas_ead, notas_presencial], labels=['EaD', 'Presencial'])


# por fim, faço minha análise. Uso prints com "f" para poder concatenar minhas variáveis
print("---------------")
print("Análise")
print("---------------")
print('')
print("Média geral:")
print(f"Ead: {media_ead_nota_geral}")
print(f"Presencial: {media_presencial_nota_geral}")

print('')
print("Mediana:")
print(f"Ead: {mediana_ead_nota_geral}")
print(f"Presencial: {mediana_presencial_nota_geral}")

print('')
print("Desvio padrão :")
print(f"Ead: {desvio_padrao_ead_nota_geral}")
print(f"Presencial: {desvio_padrao_presencial_nota_geral}")

print('')
print("Coeficiente de variação :")
print(f"Ead: {coeficiente_variacao_ead}")
print(f"Presencial: {coeficiente_variacao_presencial}")

print('----------------------------')
print("")
print("Falando por alto, podemos dizer que os alunos de EaD têm um desempenho médio e mediano superior, com menor variabilidade nas notas, enquanto os alunos presenciais apresentam maior dispersão em suas notas")
print("É notável que os dados dos alunos do EaD são mais estáveis, com menor variáção. Dá para saber isso olhando o desvio padrão e o coeficiente de variação, que são menores na modalidade EaD")
```
